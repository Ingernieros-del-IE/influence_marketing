# Data Preparation

While the data scrapper gets the data related to the instagram users, we are using two datasets from Kaggle with small variations to recreate how the algorithm will work with real data.

```python
# Import libraries
import pandas as pd
import numpy as np
import ast
```

```python
# Upload companies dataset
df_co = pd.read_csv('brands_df.csv')
df_co
```

```python
# create an ID column for the dataframe
df_co['ID_brand'] = range(1, len(df_co) + 1)
```

```python
df_co.head()
```

```python
# Upload users (influencers) dataset
df_influ = pd.read_csv('influencers_df.csv')
df_influ
```

```python
#Solve problems with apostophre
df_influ['location3'] = df_influ['location3'].str.replace('"', '')
```

```python
df_influ
```

We will now merge both dataframes, so that we can have interactions between both the companies and the users.

```python
df2_co = pd.concat([df_co]*len(df_influ), ignore_index=True)
```

```python
df2_influ = pd.DataFrame(np.repeat(df_influ.values, len(df_co), axis=0))
```

```python
# Concat both df
df = pd.concat([df2_influ, df2_co], axis=1)
```

```python
df.head()
```

```python
df.columns
```

We have a lot of columns that we don't need to create our algorithm, so we will drop them from the database.

```python
df = df.drop([1, 4, 6, 7, 8, 9, 13, 14, 'brand_name.1', 'brand_description', 'brand_post_description', 'all_descriptions', 'descriptions_translated', 'hashtags'], axis=1)
```

As we have losen the name of some columns, we will use this opportunity to rename all columns and follow the same structure.

```python
df = df.rename(columns={0: 'user_id', 2: 'user_name', 3: 'user_followers', 5: 'user_likes_mean', 10: 'user_links', 11: 'user_eng_rate', 12: 'user_hashtags_en', 15: 'user_category', 16: 'user_country', 'brand_likes': 'brand_likes_mean', 'category': 'brand_category', 'hashtags_translated': 'brand_hashtags_en', 'followers': 'brand_followers', 'location': 'brand_country', 'ID_brand': 'brand_id'})
```

```python
df.head()
```

Now, we will transform the string variable than need to be lists, remove the '' symbol and return a list of words.

```python
df['user_hashtags_en'] = df['user_hashtags_en'].apply(lambda x: ast.literal_eval(x))
df['user_country'] = df['user_country'].apply(lambda x: ast.literal_eval(x))
df['brand_hashtags_en'] = df['brand_hashtags_en'].apply(lambda x: ast.literal_eval(x))
```

```python
df
```

Calculate the different number of points (score) for each pair of user-company depending on the match of different variables.

```python
points_hashtags = (df.apply(lambda row: len(set(row['user_hashtags_en']).intersection(set(row['brand_hashtags_en']))), axis=1))*2
points_cat = np.where((df['user_category'] == df['brand_category']), 10, 0)
points_country = (df.apply(lambda row: len(set(row['brand_country']).intersection(set(row['user_country']))), axis=1))*5
points_eng = df['user_eng_rate']/df['user_followers']
df['points_hashtags'] = points_hashtags
df['points_eng'] = points_eng
df['points_cat'] = points_cat
df['points_country'] = points_country
df['points'] = df['points_cat'] + df['points_country'] + df["points_hashtags"] + df['points_eng']

df = df.drop(['points_cat', 'points_country', 'points_hashtags', 'points_eng'], axis=1)
```

```python
df
```

Check and change variables types.

```python
df.dtypes
```

```python
df['user_id'] = df['user_id'].astype(int)
df['user_name'] = df['user_name'].astype(str)
df['user_followers'] = df['user_followers'].astype(int)
df['user_likes_mean'] = df['user_likes_mean'].astype(float)
df['user_links'] = df['user_links'].astype(str)
df['user_eng_rate'] = df['user_eng_rate'].astype(float)
df['user_category'] = df['user_category'].astype(str)
df['brand_id'] = df['brand_id'].astype(int)
df['brand_name'] = df['brand_name'].astype(str)
df['brand_followers'] = df['brand_followers'].astype(int)
df['brand_likes_mean'] = df['brand_likes_mean'].astype(float)
df['brand_category'] = df['brand_category'].astype(str)
df['brand_country'] = df['brand_country'].astype(str)
df['points'] = df['points'].astype(float)
df.dtypes
```

To fit the lightfm model, we will change the name of some columns to fit the model.

```python
# Change the name of columns to fit the model.
df = df.rename(columns={"user_id": "itemID", "brand_id": "userID", "points": "rating"})
```
