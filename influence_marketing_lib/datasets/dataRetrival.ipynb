Code to get the country out of the post_Location column (locations obtained by the data scrapper in the instagram posts).

```python
geolocator = Nominatim(user_agent="my_custom_agent")
loc = []


def location(post_location):
    try:
        location = geolocator.geocode(post_location)
        if location:
            return location.address.split(',')[-1].strip()
    except (GeocoderTimedOut, GeocoderServiceError):
        return None


# Iterate over each row of the DataFrame
for index, row in df_final.iterrows():
    if pd.notnull(row['post_Location']):
        try:
            loc.append(location(row['post_Location']))
        except:
            # Handle empty response
            loc.append(None)
    else:
        loc.append(None)

# Add the country names to the DataFrame
df_final['location'] = loc
df_final
```

Code to get the hastags out of the post descriptions.

```python
def get_hashtags(text):
    if pd.isnull(text):
        return []
    hashtags = re.findall(r'#\w+', text)
    # Remove the '#' symbol from each hashtag, convert to lowercase, and store only the word in the 'has' list
    return [tag.lower()[1:] for tag in hashtags]
  
has = []
for text in df_final['post_Description']:
    if pd.isnull(text):
        has.append([])
    else:
        has.append(get_hashtags(text))

df_final['hashtags'] = has

```
