```python
import requests

def location(location_name):
    if location_name == "NaN":
        return "NaN"
    
    # Construct the URL for the geocoding API request
    url = f'https://nominatim.openstreetmap.org/search?q={location_name}&format=json'

    # Send the API request and get the response
    response = requests.get(url)

    # Parse the JSON response
    json_data = response.json()

    # Check if the results list is empty
    if not json_data:
        return "NaN"
    
    else:
        # Extract the latitude and longitude of the location
        lat = json_data[0]['lat']
        lng = json_data[0]['lon']

        # Construct the URL for the reverse geocoding API request
        url = f'https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={lng}&format=json'

        # Send the API request and get the response
        response = requests.get(url)

        # Parse the JSON response
        json_data = response.json()

        # Extract the country name from the reverse geocoding result
        country_name = json_data['address']['country']

        # return the country
        return country_name
```

```python
import re

def get_hashtags(text):
    if pd.isnull(text):
        return []
    hashtags = re.findall(r'#\w+', text)
    # Remove the '#' symbol from each hashtag, convert to lowercase, and store only the word in the 'has' list
    has = [hashtag[1:].lower() for hashtag in hashtags]
    return has
```

```python
from googletrans import Translator
from langdetect import detect

translator = Translator()
descripciones = []

for index, row in df.iterrows():
    text = row['post_Description']
    if text:
        try:
            language = detect(text)
            if language != 'en':
                translated = translator.translate(text, src=language, dest="en")
                descripcion = str(translated.text)
                descripciones.append(descripcion)
            else:
                descripciones.append(text)
        except:
            # if there is an error in translation, we can leave the original text
            descripciones.append(text)
    else:
        descripciones.append(text)

df['traducido'] = descripciones
```
